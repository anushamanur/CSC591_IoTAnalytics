```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('200256758.csv',usecols=[0])
data.rename(columns={'2m Temperature (hr. avg) (F)':'temperature'}, inplace=True)
len(data)
```

```python
df1 = data.iloc[:2197]
df2 = data.iloc[2197:]
print(len(df1), len(df2), len(data))
```

```python
# gca stands for 'get current axis'

df1.plot(kind='line', color='red', figsize=(10,4), title="Training data")
df2.plot(kind='line', color='blue', figsize=(10,4), title="Testing data")
```

```python
df1.plot(kind='hist',title='Training data')
df2.plot(kind='hist',title='Testing data')
```

```python
df1.to_csv('training.csv', sep=',')
df2.to_csv('testing.csv', sep=',')
```
